```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

load in data

```python
neat_data = np.zeros((40,3))
dqn_data = np.zeros((40,3))

with open("NEAT_training_results.csv", "r") as file:
    reader = csv.DictReader(file)
    for i, row in enumerate(reader):
        neat_data[i, 0] = row["mean"]
        neat_data[i, 1] = row["std"]
        neat_data[i, 2] = row["max"]
        
with open("DQN_training_results.csv", "r") as file:
    reader = csv.DictReader(file)
    for i, row in enumerate(reader):
        dqn_data[i, 0] = row["mean"]
        dqn_data[i, 1] = row["std"]
        dqn_data[i, 2] = row["max"]
```

making the plots

```python
y_axis = np.arange(0,2000,50)
plt.plot(y_axis, neat_data[:,0], ":", label = "NEAT mean", color="#ad15ff")
plt.plot(y_axis, neat_data[:,2], label = "NEAT max", color="#ad15ff")
plt.plot(y_axis,dqn_data[:,0], ":", label = "DQN mean", color="#ff6100")
plt.plot(y_axis,dqn_data[:,2], label = "DQN max", color="#ff6100")

plt.ylabel('Score (log10 scale)')
plt.xlabel('Games played')
plt.title('Score as a function of amount of games played')

plt.yscale("log")

plt.legend()
plt.show()
```
